```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import math
from scipy.optimize import minimize
from matplotlib.ticker import PercentFormatter
```

```python
start = "2024-01-01"
end = "2025-01-01"
```

```python
wiki_url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
table = pd.read_html(wiki_url)
tickers = table[0]['Symbol'].tolist()

```

```python
prices = yf.download(tickers,start,end,interval="1h")['Close']
# prices.dropna(axis=1,inplace=True)
prices
```

```python
normalized = prices.div(prices.iloc[0]) * 100

```

```python
correlation_matrix = normalized.corr()

abs_corr_matrix = correlation_matrix.abs()


np.fill_diagonal(abs_corr_matrix.values, 0)


max_corr = abs_corr_matrix.max().max()
tickers = np.where(abs_corr_matrix == max_corr)

ticker1 = correlation_matrix.index[tickers[0][0]]
ticker2 = correlation_matrix.columns[tickers[1][0]]

print(f"The two tickers with the highest correlation are {ticker1} and {ticker2} with a correlation of {max_corr}.")
```

```python
correlation_matrix
```

```python
corr_pairs = correlation_matrix.abs().unstack()
filtered_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]
filtered_pairs = filtered_pairs.groupby([frozenset([i, j]) for i, j in filtered_pairs.index]).max()
correlation_df = pd.DataFrame(filtered_pairs.items(), columns=['Tickers', 'Correlation'])
correlation_df['Ticker 1'] = correlation_df['Tickers'].apply(lambda x: list(x)[0])
correlation_df['Ticker 2'] = correlation_df['Tickers'].apply(lambda x: list(x)[1])
correlation_df = correlation_df.drop(columns='Tickers')
correlation_df = correlation_df.sort_values(by='Correlation', ascending=False).reset_index(drop=True)

```
